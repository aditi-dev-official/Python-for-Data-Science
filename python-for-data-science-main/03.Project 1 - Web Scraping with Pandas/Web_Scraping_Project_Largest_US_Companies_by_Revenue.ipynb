``` python

url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"

headers = {
```

# 🔎 Step-by-Step Explanation

## 1️⃣ Import Required Libraries

``` python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

-   **requests** → Sends HTTP request to fetch webpage content\
-   **pandas** → Used for data manipulation and table handling\
-   **BeautifulSoup** → Parses and navigates HTML structure

------------------------------------------------------------------------

## 2️⃣ Define the URL

``` python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"
```

This is the Wikipedia page containing the table of the largest US
companies by revenue.

------------------------------------------------------------------------

## 3️⃣ Add User-Agent Header

``` python
headers = {
    "User-Agent": "MyDataAnalysisBot/1.0 (your_email@example.com)"
}
```

Wikipedia requires a descriptive **User-Agent** to comply with their
robot policy. This identifies your script responsibly.

------------------------------------------------------------------------

## 4️⃣ Send HTTP Request

``` python
response = requests.get(url, headers=headers)
```

-   Sends request to the website
-   Stores webpage HTML content in `response`

------------------------------------------------------------------------

## 5️⃣ Parse HTML Using BeautifulSoup

``` python
soup = BeautifulSoup(response.text, 'html.parser')
```

-   Converts raw HTML into a structured object
-   Allows searching for elements like tables, divs, etc.

------------------------------------------------------------------------

## 6️⃣ Locate the Table

``` python
table = soup.find('table', class_='wikitable sortable')
```

-   Searches for the table with class `wikitable sortable`
-   This is the specific table containing company revenue data

------------------------------------------------------------------------

## 7️⃣ Convert HTML Table to Pandas DataFrame

``` python
df = pd.read_html(str(table))[0]
```

-   Converts HTML table into a structured Pandas DataFrame
-   `[0]` selects the first table found

------------------------------------------------------------------------

## 8️⃣ Display Full Table

``` python
pd.set_option('display.max_rows', 100)
print(df)
```

-   Adjusts display settings
-   Shows up to 100 rows

# 📊 Visualization (Top 10 Companies by Revenue)

After loading the data, you can create a bar chart:

``` python
import matplotlib.pyplot as plt

# Clean revenue column (if needed)
df["Revenue (USD millions)"] = (
    df["Revenue (USD millions)"]
    .replace('[\$,]', '', regex=True)
    .astype(float)
)

# Sort and select top 10
top10 = df.sort_values(by="Revenue (USD millions)", ascending=False).head(10)

plt.figure()
plt.bar(top10["Name"], top10["Revenue (USD millions)"])
plt.xticks(rotation=90)
plt.xlabel("Company")
plt.ylabel("Revenue (USD millions)")
plt.title("Top 10 US Companies by Revenue")
plt.tight_layout()
plt.show()
```

### What This Visualization Does:

-   Cleans revenue column
-   Sorts companies by revenue
-   Selects top 10
-   Creates a bar chart comparison

````python
import requests
import pandas as pd
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"

headers = {
    "User-Agent": "MyDataAnalysisBot/1.0 (fk9822647@gmail.com)"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# This is the line from your image
table = soup.find('table', class_='wikitable sortable')

# Use pandas to read the HTML string of just that specific table
df = pd.read_html(str(table))[0]

# Display all 100 rows
pd.set_option('display.max_rows', 100)

print(df)
````
